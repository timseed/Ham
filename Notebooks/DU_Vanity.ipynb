# Call Generator


Source is https://www.para.org.ph/news-guidelines-in-the-issuance-of-special-permits.html

```python
pfx=['DU','DV','DW','DY','DZ','4D','4E','4F','4G','4I']
following_char='abcdefghijklmnopqrstuvwzyz123456789'
region=3
last=[following_char[i:i+1] for i in range(0,len(following_char))]
```

```python
calls={}
cnt=0
for p in pfx:
    for l in last:
        calls[cnt]=f"{p}{region}{l}".lower()
        cnt+=1
```

```python
import re
from ham.cw import Cw
morse=Cw()

def calc_length(x):
    try:
        
        return morse.length_in_dits(morse.cw_timing(x))
    except:
        print(f"problem with {x}")
        return -1
```

```python
df_calls=pd.DataFrame.from_dict(calls,orient='index')
df_calls.columns=['call']
df_calls['last_symbol']=df_calls.call.apply(lambda x:  morse.letter_to_dot_dash(x[-1]))
df_calls['last_is_dot']=df_calls['last_symbol'].str[-1]=="."
df_calls['has_i']=df_calls['call'].str.find('i')
df_calls['has_s']=df_calls['call'].str.find('s')
df_calls['has_h']=df_calls['call'].str.find('h')

df_calls['length']=df_calls.call.apply(lambda x:  calc_length(x))
```
